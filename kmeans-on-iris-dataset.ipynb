```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn import datasets
```

```python
iris=datasets.load_iris()
x=iris.data
print(x.shape,x.dtype,"\n",x[:5])
print(iris.target_names)
print(iris.feature_names)
```

```python
from sklearn.cluster import KMeans
list1=[]

# Using Elbow method
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  y=kmeans.fit_predict(x)
  list1.append(kmeans.inertia_)
plt.plot(range(1,11),list1)
```

```python
kmeans=KMeans(n_clusters=3)
y_predict=kmeans.fit_predict(x)
print(kmeans.cluster_centers_)
#plots
print(y_predict)
%matplotlib inline
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],s=100,c='red',label='cluster 1')
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],s=100,c='blue',label='cluster 2')
plt.scatter(x[y_predict==2,0],x[y_predict==2,1],s=100,c='green',label='cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=400,marker='*',c = 'yellow', label = 'Centroid')
plt.title('Clusters of iris dataset')
plt.xlabel('features')
plt.ylabel('Target')
plt.legend()
plt.show()
```
