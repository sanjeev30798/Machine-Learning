```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import datasets
import seaborn as sns
```

```
boston=datasets.load_boston()
print(boston.feature_names,len(boston.feature_names))
print(boston.data.shape)
x=boston.data
x=x[:,5] #we have to check that which feature has high correlation value with target variable
x=x.reshape(-1,1)
y=boston.target
print(y.shape)
#print(boston.target_names)
```

```
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.17,random_state=42)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)

#train the linear model
regression=LinearRegression()
regression.fit(x_train,y_train)

y_pred=regression.predict(x_test)
y_yield=regression.predict(x_train)

print(regression.score(x_train,y_train))
print(regression.score(x_test,y_test))
print("Model underfitting")
```

```
#plots fro training dataset
plt.scatter(x_train,y_train,color="green") #original data x_train,y_train
plt.plot(x_train,y_yield,color="red") #the value obtained when x_train data is fitted
plt.xlabel("RM value")
plt.ylabel("Predicion value")
plt.title("RM value vs house price")
```

```
plt.scatter(x_test,y_test,color="green") #original x_test and y_test data
plt.plot(x_test,y_pred,color="red") #value obtained from fitting into model
plt.xlabel("RM value")
plt.ylabel("Predicion value")
plt.title("RM value vs house price")
```
